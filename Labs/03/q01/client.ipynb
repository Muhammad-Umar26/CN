```python
import socket
import os
```

```python
s = socket.socket()
```

```python
s.connect(('localhost', 9999))

while True:
    # taking reply
    reply = s.recv(1024).decode()

    if reply == "exit":
        print("Server exited, terminating connection")
        break

    _, filename, filesize = reply.split(":")
    filesize = int(filesize)

    with open("received_by_client_" + filename, "wb") as f:
        bytes_read = 0
        while bytes_read < filesize:
            data = s.recv(1024)
            f.write(data)
            bytes_read += len(data)

    print(f"File {filename} received successfully")

    # sending message
    flag = False
    while True:
        msg = input("Enter file to send file or exit to terminate: ")

        if msg == "exit":
            s.send(msg.encode())
            print("Exiting, terminating connection")
            flag = True
            break
        
        elif msg == "file":
            filepath = input("Enter file path: ")

            if not os.path.exists(filepath):
                print("File does not exist.")
                continue

            filename = os.path.basename(filepath)
            filesize = os.path.getsize(filepath)

            s.send(f"file:{filename}:{filesize}".encode())

            with open(filepath, "rb") as f:
                while True:
                    bytes_read = f.read(1024)
                    if not bytes_read:
                        break
                    s.send(bytes_read)

        break

    if flag:
        break
```
