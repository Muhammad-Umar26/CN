```python
import socket
import os
```

```python
s = socket.socket()
```

```python
s.bind(('localhost', 9999))
s.listen(5)
print("Waiting for Connection")
```

```python
c, addr = s.accept()
print(f"Got connection from {addr}")

while True:
    # sending message
    msg = input("Enter file to send file (or exit to terminate): ")

    if msg == "exit":
        c.send(msg.encode())
        print("Exiting, terminating connection")
        break

    elif msg == "file":
        filepath = input("Enter file path: ")

        if not os.path.exists(filepath):
            print("File does not exist.")
            continue

        filename = os.path.basename(filepath)
        filesize = os.path.getsize(filepath)

        c.send(f"file:{filename}:{filesize}".encode())

        with open(filepath, "rb") as f:
            while True:
                bytes_read = f.read(1024)
                if not bytes_read:
                    break
                c.send(bytes_read)

    # taking reply
    reply = c.recv(1024).decode()

    if reply == "exit":
        print("Client exited, terminating connection")
        break

    _, filename, filesize = reply.split(":")
    filesize = int(filesize)

    with open("received_by_server_" + filename, "wb") as f:
        bytes_read = 0
        while bytes_read < filesize:
            data = c.recv(1024)
            f.write(data)
            bytes_read += len(data)

    print(f"File {filename} received successfully")
```
